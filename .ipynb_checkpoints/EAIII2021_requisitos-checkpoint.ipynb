```latex
\section{Electrónica Aplicada III}


\subsection{Fundamentación}

Ante la necesidad de realizar el análisis, la síntesis e interrelación de los circuitos que componen los Sistemas de Transmisión y Recepción en RF, ver su comportamiento en detalle y el de cada una de las etapas que lo
componen, a fin de obtener una cabal comprensión de su funcionamiento.

Robustecer la enseñanza de conceptos que serán válidos con independencia de las soluciones tecnológicas particulares permitiéndole al futuro graduado adaptarse a los cambios que se gesten durante su vida profesional, la que ha de ser contemplada con un adecuado análisis de casos prácticos que permitan al cursante acortar la distancia que media entre el cálculo teórico y su concreción real.

Se considera entonces, que en la currícula haya una materia, que asocie la rama de las materias de electrónicas aplicadas, con la rama de comunicaciones, de ahí el por qué de la existencia de Electrónica Aplicada III

Carga Horaria total. Hs Reloj-128 (CIENTO VEINTIOCHO)-Hs. Cátedra-160-(CIENTO SESENTA)


\subsection{Objetivos}
\begin{itemize}
    \item  Lograr que los cursantes puedan conocer, entender el funcionamiento, analizar y proyectar los circuitos electrónicos utilizados en los sistemas de radiocomunicaciones. 
    
    \item  Tender a que el alumno canalice sus dudas, realice planteos y establezca problemáticas.
    
    \item Reconocer conceptos respecto de la problemática de la radiofrecuencia, tanto en la recepción y la transmisión de señales, identificando las funciones de cada etapa, conociendo los diagramas en bloques de los transmisores y receptores que se utilizan en diversos servicios.
    
    \item  Orientar al uso de simuladores, aún cuando  la cátedra no los imparta específicamente.
    
    \item  Desarrollar habilidades que le serán de utilidad en su vida profesional, en un clima de confianza, que permitan el trabajo en equipo y la resolución de problemáticas con respeto y valoración del pensamiento ajeno.
    
    \item  Sintetizar los conocimientos adquiridos aplicándolos al diseño de receptores y transmisores, integrando conocimientos previos de matemáticas, física, teoría de los circuitos, tecnología de los componentes, electrónica, sistemas de comunicaciones a fin de comprender la viabilidad técnico-económica de las soluciones. 
    
\end{itemize}
 
 \subsection{Contenido}

\subsubsection{Unidad temática I: Sistemas de radiocomunicación} 

\textbf{Didáctica 1:}
\begin{itemize}
    \item Introducción a los sistemas de comunicaciones en RF.
    \item Diagrama en bloques de un Transmisor y Receptor.
    \item Receptores Homodinodo y Heterodino. 
    \item Dispositivos pasivos y activos en RF.
    \item Definiciones potencia Tx (dB, dBm), sensibilidad Rx. 
\end{itemize}


\subsubsection{Unidad temática II: Circuitos de adaptación} 

\textbf{Didáctica 2:}
\begin{itemize}
    \item Teoría básica de la adaptación. 
    \item Modelo de componentes pasivos en RF. 
    \item Transformación serie paralelo e inversa. 
    \item Simulaciones SPICE.
\end{itemize}

\textbf{Didáctica 3:}
\begin{itemize}
    \item    Máxima transferencia de energía a Q constante. 
    \item Circuitos ``L'' invertida.
    \item Bobina en derivación e inductancia mutua. 
    \item Verificaciones a través de simulaciones.
\end{itemize}

\textbf{Didáctica 4:}
\begin{itemize}
    \item    Transformador sintonizado.
    \item Transformador de banda ancha. 
    \item Divisor capacitivo.
    \item Circuito ``PI''.
    \item Verificaciones a través de simulaciones.
\end{itemize}

\subsubsection{Unidad temática III: Amplificadores sintonizados}

\textbf{Didáctica 5:}
\begin{itemize}
    \item Modelo de dispositivos activos para alta frecuencia.
    \item El amplificador simple sintonizado: ecuaciones de transferencia y diagrama de polos y ceros.
\end{itemize}

\textbf{Didáctica 6:}
\begin{itemize}
    \item El amplificador simple sintonizado: ancho de banda, aproximación de banda angosta. Producto ganancia por ancho de banda.
    \item Amplificador multietapa sincrónico
    \item  Verificaciones a través de simulaciones.
\end{itemize}

\textbf{Didáctica 7:}
\begin{itemize}
    \item El amplificador doble sintonizado: ecuaciones de transferencia, diagrama de polos y ceros, aproximación de Banda Angosta, Transferencia y Ancho de Banda. 
    \item Topologías de pasa-bandas Sincrónicos y Butterwoth.
    \item Verificaciones a través de simulaciones.
\end{itemize}

\textbf{Didáctica 8:}
\begin{itemize}
    \item Análisis de la estabilidad del amplificador.
    \item Factores de Stern y Linvill. 
\end{itemize}

\textbf{Didáctica 9:}
\begin{itemize}
    \item Conceptos de Estabilidad, círculos de estabilidad y su ubicación en el diagrama de Smith, factor de estabilidad de Rollett. 
    \item Ganancia de transducción. 
    \item Desadaptación, Neutralización y Unilateralización. 
    \item Verificaciones a través de simulaciones.
\end{itemize}

 
\subsubsection{Unidad temática IV: Ruido eléctrico} 

\textbf{Didáctica 10:}
\begin{itemize}
    \item Definiciones y terminología en el estudio del ruido: Relación señal a ruido, Ancho de banda equivalente, Cifra de ruido y Temperatura de ruido. 
    \item Ruido térmico en resistencias y redes. 
    \item Ruido en dispositivos activos. 
\end{itemize}

\textbf{Didáctica 11:}
\begin{itemize}
    \item Amplificadores de bajo ruido (LNA)
    \item Consideraciones de la influencia del ruido en el diseño de amplificadores. 
    \item  Ancho de banda de ruido equivalente.
    \item  Elección de la resistencia óptima del generador desde el punto de vista de ruido. 
    \item  Verificaciones empleando simulaciones.
\end{itemize}

\subsubsection{Unidad temática V: Detección en A.M.} 

\textbf{Didáctica 12:}
\begin{itemize}
    \item  Detector de envolvente, circuito a diodo, rendimiento, resistencia equivalente de entrada,  distorsiones. 
    \item  Control automático de ganancia mediante sistema directo e inverso. 
    \item  Verificaciones empleando simulaciones.
\end{itemize}

\subsubsection{Unidad temática VI: Amplificadores de potencia sintonizados} 

\textbf{Didáctica 13:}
\begin{itemize}
    \item Consideraciones generales de transmisores: modelos de dispositivos activos, filtros. 
    \item Comparación entre las clases de amplificadores ``A'' / ``AB'' / ``B'' / ``C''. 
    \item El amplificador ``Clase C''.
\end{itemize}

\textbf{Didáctica 14:}
\begin{itemize}
    \item  El amplificador ``Clase C'' práctico.
    \item  Verificaciones empleando simulaciones.
\end{itemize}

\textbf{Didáctica 15:}
\begin{itemize}
    \item Distorsiones armónicas en amplificadores ``Clase C''.
    \item Verificaciones empleando simulaciones.
\end{itemize}

\textbf{Didáctica 16:}
\begin{itemize}
    \item    Amplificador de RF Modulado. 
    \item  Verificaciones empleando simulaciones.
\end{itemize}

\subsubsection{Unidad temática VII: Osciladores sinusoidales}

\textbf{Didáctica 17:}
\begin{itemize}
    \item Generalidades de los Osciladores y su clasificación. 
    \item Osciladores de resistencia negativa. 
    \item Verificaciones a través de simulaciones.
\end{itemize}

\textbf{Didáctica 18:}
\begin{itemize}
    \item    Osciladores Colpitts.
    \item    Osciladores Clapp.
    \item Verificaciones a través de simulaciones.
\end{itemize}

\subsubsection{Unidad temática VIII: Mezcladores}

\textbf{Didáctica 19:}
\begin{itemize}
    \item Teoría del mezclador y su análisis espectral. 
    \item Terminología utilizada en mezcladores. 
    \item Ganancia de conversión, nivel de compresión, rango dinámico. 
    \item Mezcladores a diodo y mezcladores balanceados.
    \item Verificaciones a través de simulaciones.
\end{itemize}

\textbf{Didáctica 20:}
\begin{itemize}
    \item Mezcladores y Conversores con dispositivos activos: FET, Bipolar, MOSFET de doble compuerta. 
    \item Mezcladores empleando celda de Gilbert.
    \item Verificaciones a través de simulaciones
\end{itemize}

\subsubsection{Unidad temática IX: Moduladores y demoduladores angulares.} 

\textbf{Didáctica 21:}
\begin{itemize}
    \item Modulación angular. 
    \item Generación de F.M.: directa e indirecta.
    \item Multiplicadores y heterodinaje. 
    \item Transistor de reactancia.
    \item Verificaciones a través de simulaciones
\end{itemize}

\textbf{Didáctica 22:}
\begin{itemize}
    \item  Modulador Beleskas, Armstrong, diodo varicap. 
    \item  Verificaciones a través de simulaciones
\end{itemize}

\textbf{Didáctica 23:}
\begin{itemize}
    \item Características del detector de FM.
    \item Discriminador por desplazamiento de fase, discriminador Foster–Seeley, detector de razón, detector de cuadratura. 
    \item Verificaciones a través de simulaciones
\end{itemize}

\subsubsection{Unidad temática X: Lazos de fijación de fase.} 

\textbf{Didáctica 24:}
\begin{itemize}
    \item Esquema simplificado de la operación del lazo. 
    \item Diagrama en bloques de la operación de un PLL. 
    \item Análisis y terminología utilizada en PLL. 
    \item El oscilador de lazo (VCO). 
\end{itemize}


\subsection{Metodología}

Modalidades de enseñanza empleadas según tipo de actividad (teórica-práctica). 
Las clases serán dictadas basándose en el método inductivo deductivo, con exposición teórico-práctica de cada una de las Unidades Temáticas, siempre en función de las posibilidades de tiempo y disponibilidad.
Las exposiciones didácticas de los diversos temas se realizarán para recalcar los puntos fundamentales de cada tema. Si bien se intentara cubrir todos los temas del programa, es responsabilidad del alumno conocer los temas del mismo. 
Los encuentros pueden ser en un aula o mediante videoconferencias, también dependiendo de disponibilidad.
En caso de ser mediante videoconferencia, cada uno de los alumnos deberá contar para cada encuentro con una computadora, o similar, con conexión a internet estable, cámara y micrófono. 

El planteo de ejemplos, problemas y/o proyectos está diseñado para que, en etapas sucesivas, el cursante pueda enfrentar la problemática de la RF, la que a diferencia de lo estudiado en otras asignaturas implica: procesar señales pequeñas en presencia de interferencias fuertes, procesar señales con elementos cuyos parámetros cambian durante el ciclo de la señal, evaluar alinealidades introducidas por componentes activos y pasivos. 

Como complemento a los ejercicios, se realizaran comprobaciones mediante simulares de software como Qucs, LTSpice y Smith V3.10. Dichas comprobaciones consisten en verificar los conocimientos teóricos adquiridos.

\subsubsection{Consultas}

Las dudas relacionadas con ejercicios, teoría o proyecto pueden ser planteadas en el aula, en el foro del aula virtual o en algún otro medio de comunicación establecido para el curso. 
Se busca una participación conjunta y grupal del aprendizaje, por tal motivo los alumnos deberán tratar de responder a las cuestiones planteadas. Los profesores realizan un seguimiento de cada consulta, donde su intervención sera para ayudar a entender un concepto o realizar alguna aclaración o si la cuestión no esta debidamente respondida. Si la cuestión no es respondida por los alumnos, se intentara dar respuesta en el tiempo del encuentro. 
```
